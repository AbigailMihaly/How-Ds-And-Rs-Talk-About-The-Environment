Import .csv files from the word_frequency_analysis notebook

```python
df_Rhouse = pd.read_csv('top_R_house_phrases.csv')
df_Dhouse = pd.read_csv('top_D_house_phrases.csv')
df_Rsenate = pd.read_csv('top_R_senate_phrases.csv')
df_Dsenate = pd.read_csv('top_D_senate_phrases.csv')
```

### Merge the House and Senate dataframes for each party ###

```python
#DEMOCRATS

# create an empty list of rows

D_rows = []
```

```python
# fill the rows 

senate_only_count = 0
house_only_count = 0
duplicate_count = 0

for i in range(0, 150): # run through each row in the house dataframe

    # locate the noun phrase in that row
    phrase = df_Dhouse.iloc[i]['Noun Phrase']

    if phrase in df_Dsenate['Noun Phrase'].values:  # check if the noun phrase in the SENATE list as well as the HOUSE
        print(f'{phrase} is on both the House and Senate lists.') # if so, print that it is in both
        duplicate_count = duplicate_count + 1 # count up the number of duplicate phrases

        # define information for a new row in new merged dataframe
        
        # locate both frequency numbers
        Dhouse_frequency = df_Dhouse.iloc[i]['Frequency'].item() # in the House df row we're in, find Frequency
        Dsenate_frequency = df_Dsenate.loc[df_Dsenate['Noun Phrase'] == phrase, 'Frequency'].item() # in the Senate df row with the same noun phrase, find its frequency in that df

        # and ad the frequencies together
        Dtotal_frequency = Dhouse_frequency + Dsenate_frequency
    
    # if noun phrase on HOUSE list but NOT SENATE list.
    # add the information into a new row
    elif phrase not in df_Dsenate['Noun Phrase'].values:
        print(f'{phrase} is only on House list.')
        house_only_count = house_only_count + 1
    
        # define information for a new row in new merged dataframe
        Dhouse_frequency = df_Dhouse.iloc[i]['Frequency'].item() # in the House df row, find Frequency in the D House list
        Rsenate_frequency = 0
        Dtotal_frequency = Dhouse_frequency

    # now create the row from the above info
    new_row = {
        'noun_phrase': phrase,
        'Dtotal_frequency': Dtotal_frequency,
        'Dhouse_frequency': Dhouse_frequency,
        'Dsenate_frequency': Dsenate_frequency
    }
    
    D_rows.append(new_row)
```

```python
# Also need to add in the noun phrases on the SENATE but not HOUSE lists:
for i in range(0, 150): # run through the rows in the Senate dataframe

    # locate the Noun Phrase in the row were in 
    phrase = df_Dsenate.iloc[i]['Noun Phrase']
    
    # if noun phrase on SENATE list but NOT HOUSE list:
    if phrase not in df_Dhouse['Noun Phrase'].values:
        print(f'{phrase} is only on Senate list.') # print it out
        senate_only_count = senate_only_count + 1 # count up how many phrases are senate only

        # define information for a new row in new merged dataframe
        Dhouse_frequency = 0 
        Dsenate_frequency = df_Dsenate.iloc[i]['Frequency'].item()
        Dtotal_frequency = Dsenate_frequency

        # create a new column in the merged dataframe with the phrase
        new_row = {
        'noun_phrase': phrase,
        'Dtotal_frequency': Dtotal_frequency,
        'Dhouse_frequency': Dhouse_frequency,
        'Dsenate_frequency': Dsenate_frequency
    }   
        D_rows.append(new_row)
```

### Now make a new column in each of the the dataframes with the category, to enable sorting ###

First, sort phrases into buckets

```python
# Define a function to categorize each individual item

def assign_phrase_category(item):
    if item in not_interesting:
        return "not interesting"
    elif item in places:
        return "place"
    elif item in lawmakers:
        return "lawmaker"
    elif item in federal_actors:
        return "actor"
    elif item in legislation_and_initiatives:
        return "legislation"
    else:
        return "general"
```

```python
# Define phrase category

df_D['phrase_category'] = df_D['noun_phrase'].apply(assign_phrase_category)
df_R['phrase_category'] = df_R['noun_phrase'].apply(assign_phrase_category)
```

```python
# print out how many of each category to make sure it worked

print(df_D['phrase_category'].value_counts())
print(df_R['phrase_category'].value_counts())
```

Create .csvs with each of the categories

```python
#Democrats

general = df_D[df_D['phrase_category'] == 'general'].sort_values(by='Dtotal_frequency', ascending=False)
places = df_D[df_D['phrase_category'] == 'place'].sort_values(by='Dtotal_frequency', ascending=False)
lawmakers = df_D[df_D['phrase_category'] == 'lawmaker'].sort_values(by='Dtotal_frequency', ascending=False)
federal_actors = df_D[df_D['phrase_category'] == 'federal_actor'].sort_values(by='Dtotal_frequency', ascending=False)
leg_and_initiatives = df_D[df_D['phrase_category'] == 'legislation_and_initiatives'].sort_values(by='Dtotal_frequency', ascending=False)


general.to_csv('D_top_generalphrases.csv')
places.to_csv('D_top_placephrases.csv')
lawmakers.to_csv('D_top_lawmakerphrases.csv')
federal_actors.to_csv('D_top_actorphrases.csv')
leg_and_initiatives.to_csv('D_top_legandinitiatives.csv')
```
